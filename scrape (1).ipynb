```python
def remove_commas(value):
    if isinstance(value, str):
        return value.replace(',', '')
    elif isinstance(value, (int, float)):
        return str(value).replace(',', '')
    else:
        return value

# Apply the function to all columns
df.iloc[:, 2:] = df.iloc[:, 2:].applymap(remove_commas)

# Convert the columns back to numeric (if needed)
df.iloc[:, 2:] = df.iloc[:, 2:].apply(pd.to_numeric, errors='coerce')

df
```

```python
category_list = ['Business travel', 'Crew spending',
       'Other business travel1', 'Personal travel', 'Health related',
       'Education related', 'Other personal travel1']


for trade_type in list(df.Trade.unique()):
    
    sum_all = df.loc[(df.Travel_Category == 'Travel, total') & (df.Trade == str(trade_type))].values[0][2:]
    sum_others = df.loc[(df.Travel_Category.isin(category_list)) & (df.Trade == trade_type)].sum()[2:]
    
    sum_all_float = [float(val) for val in sum_all]
    sum_others_float = [float(val) for val in sum_others]
    difference = list(map(lambda x, y : x-y, sum_all_float, sum_others_float))
    new_row = [trade_type, "Others"] +  difference
    df.loc[len(df)] = new_row
```
